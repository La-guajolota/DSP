# ** Señales Continuas ** 

```python
# Parte 1 - Importar librerías:

import numpy as np                      # Librería para analizar números.
import matplotlib.pyplot as plt         # Librería para gráficas. 
```

```python
# Parte 2 - Configuraciones iniciales: 
w = 1                                   # variable a utilizar.
fx = lambda t: np.sin(w*t)              # Funciones vacías, una sola cosa.

# Creación del intervalo de tiempo: 
t0 = 0                                  # Tiempo inicial.
tn = 2*np.pi                            # Tiempo final.
muestras = 51                           # Número de muestras.

# Parte 3 - Procedimiento:
ti = np.linspace(t0,tn,muestras)        # Valor inicial, valor final y número de valores ---> linspace
senal = fx(ti)                          
```

```python
# Parte 4 - Salida y gráfica:
np.set_printoptions(precision=4)        
print("tiempo: ")
print(ti)                           # Imprime el tiempo.
print("señal: x(t)")    
print(senal)                        # Imprime la señal.

# Gráfica: 
plt.figure()
plt.plot(ti,senal)                  # Valores que se imprimirán.
plt.xlabel('t [s]')                 # Valores en el label x.
plt.ylabel('Señal x(t)')            # Valores en el label y. 
plt.show()                          # Muestra gráfica. 
```

# ** Señal Discreta **


```python
w = 2 * np.pi / 12
fx = lambda n: np.sin(w*n)
```

```python
# Parte 3 - Procedimiento: 
# Creación de vector de N muestras: 
n0 = 0
muestras = 12 + 1

ni = np.arange(n0,n0+muestras,1)
senal_discreta = fx(ni)
   
```

```python
# Salida y gráficas: 
np.set_printoptions(precision=4)        
print("n: ")
print(ni)                           
print("señal: x(n)")    
print(senal_discreta)               # Imprime la señal discreta.

# Gráfica: 
plt.figure()
plt.stem(ni,senal_discreta)         # Valores que se imprimirán.
plt.xlabel('n')                     # Valores en el label x.
plt.ylabel('Señal discreta x[n]')   # Valores en el label y. 
plt.show()                          # Muestra gráfica.
```

# ** Sucesiones Unitarias **

```python
# Señal 1 - Impulso unitario: 

x1 = np.array([0,0,0,1,0,0,0,0])            # Creación de arreglo con 8 valores.
n1 = np.arange(-4,4)                        

plt.figure()                                # Creación de la figura.
plt.stem(n1,x1)                             # Valores que tendrá la gráfica.
plt.xlabel('n')                             # Nombre del label x.
plt.ylabel('Amplitud')                      # Nombre del label y.
plt.title('Impulso Unitario')               # Título de la gráfica.
plt.show()                                  # Muestra la gráfica.
```

```python
# Señal 2 - Escalón unitario: 

x2 = np.concatenate((np.zeros(10),np.ones(11))) # Creación de vector rápido concatenando valores. 
n2 = np.arange(-10,11)                          

plt.figure()                                    # Creación de la figura.
plt.stem(n2,x2)                                 # Valores que tendrá la gráfica.
plt.xlabel('n')                                 # Nombre del label x.
plt.ylabel('Amplitud')                          # Nombre del label y.
plt.title('Escalón Unitario')                   # Título de la gráfica.
plt.show()                                      # Muestra la gráfica.
```

```python
# Señal 3 - Sucesión exponencial
a = 0.8
n3 = np.arange(-5,21)
x3 = np.concatenate((np.zeros(5), a**np.arange(21)))
plt.figure()                                    # Creación de la figura.
plt.stem(n3,x3)                                 # Valores que tendrá la gráfica.
plt.xlabel('n')                                 # Nombre del label x.
plt.ylabel('Amplitud')                          # Nombre del label y.
plt.title('Exponencial')                        # Título de la gráfica.
plt.show()                                      # Muestra la gráfica.
```
